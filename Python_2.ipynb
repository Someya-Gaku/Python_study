```python
# Scipyの基礎 // P45
# import scipy as sp                         # scipyのimport
import scipy.linalg as linalg                # 線形代数用のライブラリ(linalg)
from scipy.optimize import minimize_scalar   # 最適化計算(最小値)用の関数
                                             # → scipy.optimize.minimize_scalarを簡易化している
```

```python
# scipyを用いた行列計算  // P45

matrix = np.array([[1, -1, -1],[-1, 1, -1],[-1, -1, 1]])

print ("行列式:\n", linalg.det(matrix))      # .det() → 行列式の計算
print ("---")
print ("逆行列:\n", linalg.inv(matrix))      # .inv() → 逆行列の計算
print ("---")

# ある行列Aと、その行列Aの逆行列を掛けた行列は、単位行列になる → 正しいか確認できる
print (matrix.dot(linalg.inv(matrix)))       # .dot() → 行列同士の計算 (P43で既出)
```

```python
# 固有値と固有ベクトルの計算 // P46

linalg.eig(matrix)                            # .eig() → 固有値と固有ベクトルを求める

# 変数A , 変数B = linalg.eig(matrix) とすると、Aに固有値[-1.+0.j,  2.+0.j,  2.+0.j]、Bに固有ベクトル[[ 0.577,  0.816,  0.408], が入る
#                                                                                                    [ 0.577, -0.408, -0.816],
#                                                                                                    [ 0.577, -0.408,  0.408]]]
```

```python
# ニュートン法（newton関数） // P47
# 以下の方程式の解を求めることを考える。
```



$f(x) = x^2 + 2x + 1$  (答えは -1)

```python
# まずは上記の関数を定義する // 定義についてはP23

def my_function(x):
  return (x**2 + 2*x + 1)                # x^2 + 2x + 1ということ

from scipy.optimize import newton        # ニュートン法の読み込み

print (newton(my_function, 0))           # newton() → 引数を２つとる。1つ目に元の式(定義した関数 my_function)、2つ目に解を求める条件( f(x) = 0 の "0" ) をセットしている

# ↓数値計算による誤差が発生している
```

```python
# この関数における最小値を求める
#from scipy.optimize import minimize_scalar   # 最適化計算(最小値)用の関数
                                              # → scipy.optimize.minimize_scalarを簡易化している

print(minimize_scalar(my_function, method = "Brent"))
# パラメータ method の 指定している "Brent"とは、Brent法を指している。（数値解析における求根アルゴリズムの1つ）

# message:
#         Optimization terminated successfully;                     → 最適化は正しく終了しました。
#         The returned value satisfies the termination criteria     → 返却値は終了値を満たしています
#         (using xtol = 1.48e-08 )                                  → xtol(求めるxの値の精度) = 1.48e-8 を使いました。
```

```python
# 総合問題 2-1
# モンテカルロ法

import math

N=10000
x = np.random.uniform(0.0, 1.0, N)
y = np.random.uniform(0.0, 1.0, N)
```

```python
# 総合問題 2-1-2_1

#円の中に入ったxとy
inside_x = []
inside_y = []

#円の外に出たxとy
outside_x = []
outside_y = []

count_inside = 0
for count in range(0,N):
    d = math.hypot(x[count],y[count])       # x-y座標の原点からx,yの点の距離を計算
    if d < 1:
        count_inside += 1
        # 円の内部に入った時のxとyの組み合わせ
        # appendはリストに要素を追加するメソッド
        inside_x.append(x[count])
        inside_y.append(y[count])

    else:
        # 円の外に出たときのxとyの組み合わせ
        outside_x.append(x[count])
        outside_y.append(y[count])

print("円の内部に入った数:", count_inside)
```

```python
# 総合問題 2-1-2_2

# 図のサイズ
plt.figure(figsize = (5,5))

# 円を描くためのデータ
circle_x=np.arange(0,1,0.001)
circle_y = np.sqrt(1-circle_x*circle_x)

plt.plot(circle_x, circle_y)

# 円の中に入っているのが red
plt.scatter(inside_x, inside_y, color = 'r')
# 円の外に出てるのが blue
plt.scatter(outside_x, outside_y, color='b')

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```
