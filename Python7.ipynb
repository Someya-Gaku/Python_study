```python
# 棒グラフ // P185
# 表示するデータ
x = [1, 2, 3]
y = [10, 1, 4]

# グラフの大きさ指定
plt.figure (figsize = (10, 6))

# 棒グラフの設定
plt.bar(x, y, align = "center", width = 0.5)


# ラベル
plt.xticks(x, ["A Class", "B Class", "C Class"])
plt.xlabel("Class")
plt.ylabel("score")

plt.grid(True)
```

```python
# 横の棒グラフ // P185

# 表示するデータ
x = [1, 2, 3]
y = [10, 1, 4]

# グラフの大きさ指定
plt.figure (figsize = (10, 6))

# 横の棒グラフの設定
plt.barh(x, y, align = "center")

# ラベル
plt.yticks(x, ["A Class", "B Class", "C Class"])  # ".xticks()" ではなく ".yticks()"
plt.ylabel("Class")
plt.xlabel("score")

plt.grid(True)
```

```python
# 複数のグラフを描く // P186

# データの準備
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

# X軸のデータ
x = np.arange(len(y1))

# グラフの幅
w = 0.4

# 大きさ
plt.figure(figsize = (10, 6))

# グラフ(y2は、"+w"をして右にずらす)
plt.bar(x, y1, color = "blue", width = w, label = "Math first", align = "center")
plt.bar(x+w, y2, color = "green", width = w, label = "Math final", align = "center")

# 凡例を最適な位置に配置
plt.legend (loc = "best")

plt.xticks(x + w / 2, ["Class A", "Class b", "Class C"])
plt.grid(True)
```

```python
# 積み上げ棒グラフ // P186

# データの準備
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

# X軸
x = np.array([1, 2, 3, 4, 5])

# 大きさ
plt.figure(figsize = (10, 6))

# グラフ
p1 = plt.bar(x, height1, color = "blue")
p2 = plt.bar(x, height2, bottom = height1, color = "lightblue")

# 凡例表示
plt.legend ((p1[0], p2[0]), ("Class 1", "Class 2"))
```

```python
# 円グラフ // P187

labels = ["Frpgs", "hogs", "Dogs", "Logs"]
sizes = [15, 30, 45, 10]
colors= ["yellowgreen", "gold", "lightskyblue", "lightcoral"]
explode = (0, 0.1, 0, 0)

# 大きさ・凡例
plt.figure(figsize = (15, 6))
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 90)

# 円を丸く描画
plt.axis("equal")
```

# 練習問題 - 解答 (7-1 ～ 7-3) / P190

**練習問題 7-1**

数学の成績を示すデータである「student-mat.csv」を使って学校を選んだ理由(reason) を円グラフ化して、それぞれの割合を出しなさい

```python
import requests
import zipfile
from io import StringIO
import io

zip_file_url = "http://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip"

r = requests.get(zip_file_url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()

#データの読み込み
student_data_math = pd.read_csv("student-mat.csv", sep=";")
```

```python
# 7-1解答

student_data_math.groupby("reason").size().plot(kind = "pie", autopct = "%1.1f%%", startangle = 90)

plt.ylabel("")
plt.xlabel("equal")
```

**練習問題 7-2**

7-1と同じデータで、higher(高い教育を受けたいかどうか。 "yes" or "no" ) を軸にして、それぞれの数学の成績(G3)の平均値を棒グラフで表示しなさい。また、そこから何か推測しなさい。

```python
# 7-2解答

student_data_math.groupby("higher")["G3"].mean().plot(kind = "bar")
plt.xlabel("higher")
plt.ylabel("G3 grade avg")

# "YES (高い教育を受けたい人)" の方が "G3 (数学の成績)" が高い傾向があると推測できる
```

**練習問題 7-3**

7-1と同じデータで、traveltime (通学時間) を軸にして、それぞれの数学の成績(G3)の平均値を横向きの棒グラフで表示しなさい。また、そこから何か推測しなさい。

```python
# 7-3解答

student_data_math.groupby(["traveltime"])["G3"].mean().plot(kind = "barh")
plt.xlabel("G3 grade avg")

# "traveltime (通学時間)" が長い方が "G3 (数学の成績)" が低い傾向にあると推測できる
```

# 応用：金融データの可視化 / P191
(スキップ可 / 練習問題無し)

```python
# 可視化する金融データ // P191

# 日付データの設定 (freq = "T" で1分ごとにデータを生成する)
idx = pd.date_range("2015/01/01", "2015/12/31 23:59", freq = "T")

# 乱数(1 or -1)の生成
dn = np.random.randint(2, size = len(idx)) * 2-1

# ランダムウォーク(ランダムに数値が増減するデータ)の作成
rnd_walk = np.cumprod(np.exp(dn * 0.0002)) * 100          # .cumprod → 累積積の算出 (要素A * 要素B * 要素C * ・・・と和が累積されていく) // P191

# resample("B")でデータを営業日単位でリサンプリング
# ohlcメソッドで「open」「high」「low」「close」の4つのデータにする。
df = pd.Series(rnd_walk, index = idx).resample("B").ohlc()
```

```python
df.plot(figsize = (15, 6), legend = "best", grid = True)
```

```python
# ローソクチャートを表示するライブラリ(Plotly) // P192
from plotly.offline import init_notebook_mode, iplot
from plotly import figure_factory as FF

# ↓Jupyter notebookを使っている人用(コメントを外してください)
# init_notebook_mode(connected = True)

# ローソクチャートの設定
fig = FF.create_candlestick(df.open, df.high, df.low, df.close, dates = df.index)
iplot(fig)
```
